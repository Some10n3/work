```python
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
# Define the distance matrix as a DataFrame for easier manipulation
labels = ['A', 'B', 'C', 'D', 'E', 'F']
distance_matrix = np.array([
    [0, 8, 20, 16, 12, 4],
    [8, 0, 23, 21, 13, 7],
    [20, 23, 0, 5, 20, 18],
    [16, 21, 5, 0, 13, 17],
    [12, 13, 20, 13, 0, 6],
    [4, 7, 18, 17, 6, 0]
])
distance_df = pd.DataFrame(distance_matrix, index=labels, columns=labels)
distance_df
```

```python
# Perform Single Linkage clustering
single_linkage = sch.linkage(distance_matrix[np.triu_indices_from(distance_matrix, 1)], method='single')

# Plot dendrogram for Single Linkage
plt.figure(figsize=(10, 6))
sch.dendrogram(single_linkage, labels=labels, distance_sort='ascending')
plt.title('Hierarchical Clustering Dendrogram (Single Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
# Perform Average Linkage clustering
average_linkage = sch.linkage(distance_matrix[np.triu_indices_from(distance_matrix, 1)], method='average')

# Plot dendrogram for Average Linkage
plt.figure(figsize=(10, 6))
sch.dendrogram(average_linkage, labels=labels, distance_sort='ascending')
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as sch
from scipy.spatial.distance import squareform
import matplotlib.pyplot as plt

# Define the distance matrix (2D) as per the provided image
labels = ['A', 'B', 'C', 'D', 'E', 'F']
distance_matrix = np.array([
    [0, 8, 20, 16, 12, 4],
    [8, 0, 23, 21, 13, 7],
    [20, 23, 0, 5, 20, 18],
    [16, 21, 5, 0, 13, 17],
    [12, 13, 20, 13, 0, 6],
    [4, 7, 18, 17, 6, 0]
])

# Convert the 2D distance matrix to a condensed 1D format using squareform
condensed_dist_matrix = squareform(distance_matrix)

# Perform Single Linkage clustering
single_linkage = sch.linkage(condensed_dist_matrix, method='single')

# Plot dendrogram for Single Linkage
plt.figure(figsize=(10, 6))
sch.dendrogram(single_linkage, labels=labels, distance_sort='ascending')
plt.title('Hierarchical Clustering Dendrogram (Single Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Perform Average Linkage clustering
average_linkage = sch.linkage(condensed_dist_matrix, method='average')

# Plot dendrogram for Average Linkage
plt.figure(figsize=(10, 6))
sch.dendrogram(average_linkage, labels=labels, distance_sort='ascending')
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
